```python
import numpy as np
import pandas as pd
# import matplotlib as plt
from matplotlib import pyplot as plt
import seaborn as sns
```

```python
data = pd.read_csv("Dataset.csv")
data = data.iloc[4:]

# for x in data.columns:
data["Region"] = data["Region"].str.replace("\n","")
data["Name"] = data["Name"].str.replace("\n","")
data["University"] = data["University"].str.replace("\n","")
```

```python
plt.figure(figsize=(30,15))
sns.boxplot(x="Region", y="Fake article count",
                 data=data).set_title('Count of Fake Article by Region',fontsize=30)
```

```python
plt.figure(figsize=(30,15))
sns.boxplot(x="Region", y="Fake citation count",
                 data=data).set_title('Count of Fake Citations by Region',fontsize=30)
```

```python
plt.figure(figsize=(30,10))
sns.boxplot(x="Region", y="Percent of Fake Articles",
                 data=data).set_title('Percentage of Fake Article by Region',fontsize=30)
```

```python
plt.figure(figsize=(30,10))
sns.boxplot(x="Region", y="Percent of Fake Citations",
                 data=data).set_title('Percentage of Fake Citations by Region',fontsize=30)
```
